```python
import matplotlib.pyplot as plt
import cv2

# Load image
img = cv2.imread('/content/drive/MyDrive/PCVK/853e8262213fb40c519991ac4e2cc084.jpg')
plt.figure(figsize=(10, 10))  # Mengatur ukuran gambar
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Menampilkan channel Red-Blue
img_rb = img.copy()
img_rb[:, :, 1] = 0
plt.imshow(cv2.cvtColor(img_rb, cv2.COLOR_BGR2RGB))
plt.title('Red-Blue')
plt.show()
```

```python
# Menampilkan channel Green-Blue
img_gb = img.copy()
img_gb[:, :, 2] = 0
plt.imshow(cv2.cvtColor(img_gb, cv2.COLOR_BGR2RGB))
plt.title('Green-Blue')
plt.show()
```

```python
cropped_img = img[20:115, 25:120]
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_red = img[5:30, :, 2]
plt.imshow(img_red, cmap='gray')
plt.show()
```

```python
import numpy as np

img_with_boxes = np.zeros((500, 500, 3), dtype=np.uint8)

for _ in range(5):
    pt1 = tuple(np.random.randint(0, 500, 2))
    pt2 = tuple(np.random.randint(0, 500, 2))
    color = tuple(np.random.randint(0, 256, 3).tolist())
    img_with_boxes = cv2.rectangle(img_with_boxes, pt1, pt2, color, -1)

plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.title('Random')
plt.show()
```

```python
flipped_img = cv2.flip(img, -1)  # Flip vertikal dan horizontal
plt.imshow(cv2.cvtColor(flipped_img, cv2.COLOR_BGR2RGB))
plt.title('Flipped')
plt.show()
```
