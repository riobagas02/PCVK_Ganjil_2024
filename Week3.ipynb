```python
print('Mengubah tingkat kecerahan citra')
print('---------------------------------')

try:
  brightness = int(input(' Masukkan nilai kecerahan : '))
except ValueError:
  print('Error, not a number')

original = cv.imread('/content/drive/MyDrive/PCVK/53a1868f14b8bf6330679f2f81694369.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y, x, c] = np.clip(original[y, x, c] + brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```python
import cv2 as cv
import numpy as np
from google.colab.patches import cv2_imshow

print('Mengubah inversi citra')
print('-----------------------')

try:
    brightness = int(input('Masukkan nilai kecerahan: '))
except ValueError:
    print('Error, bukan angka')

# Membaca citra asli
original = cv.imread('/content/drive/MyDrive/PCVK/Screenshot 2024-09-14 150925.png')

inverse_image = np.zeros(original.shape, original.dtype)

# Menerapkan inversi citra
for y in range(original.shape[0]):
    for x in range(original.shape[1]):
        for c in range(original.shape[2]):
            inverse_image[y, x, c] = 255 - brightness_image[y, x, c]

# Menggabungkan citra asli, citra dengan kecerahan diubah, dan citra inversi
final_frame = cv.hconcat((original, inverse_image))

# Menampilkan hasil
cv2_imshow(final_frame)
```
